# Simple Linear Regression

```python
# import librarirs
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Read the CSV file into a DataFrame
df = pd.read_csv('../Dataset/student_placement.csv')
```

```python
# Show the first 5 rows of the DataFrame
df.head()
```

```python
# Create a scatter plot between CGPA and package
plt.scatter(df['cgpa'], df['package'])
# Label the x-axis as 'cgpa'
plt.xlabel('cgpa')
# Label the y-axis as 'package(in lpa)'
plt.ylabel('package(in lpa)')
```

```python
# Select the first column (cgpa) as feature X
X = df.iloc[:, 0:1]
# Select the last column (package) as target y
y = df.iloc[:, -1]
```

```python
# Display the feature values (cgpa column)
X
```

```python
# Split the data into training and testing sets
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y, test_size=0.2, random_state=2)
```

```python
# Import the LinearRegression model from scikit-learn
from sklearn.linear_model import LinearRegression
```

```python
# Create an object (model) of the LinearRegression class
lr = LinearRegression()
```

```python
# Train the Linear Regression model using the training data
lr.fit(X_train, y_train)
```

```python
# Display the test feature data (cgpa values used for testing)
X_test
```

```python
# Predict the package for the 3rd cgpa value in the test set
# .iloc[2] selects the 3rd value, and .values.reshape(1,1) converts it into the required 2D array
lr.predict(X_test.iloc[2].values.reshape(1, 1))
```

```python
# Plot actual data points
plt.scatter(df['cgpa'], df['package'])

# Plot regression line
plt.plot(X_train, lr.predict(X_train), color='red')

# Label axes
plt.xlabel('cgpa')
plt.ylabel('package(in lpa)')
```

```python
# Get the slope (coefficient) of the regression line
m = lr.coef_
m
```

```python
# Get the intercept of the regression line
b = lr.intercept_
b
```

```python
# Using the regression equation y = m*x + b to predict package for cgpa = 8.58
m * 8.58 + b
```

```python
# Predict the package when cgpa = 9.5
m * 9.5 + b
```

```python
# Predict the package when cgpa = 10
m * 10 + b
```
