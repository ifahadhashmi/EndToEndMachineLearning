# Working with SQL

```python
# pip install mysql.connector
import mysql.connector
```

```python
# Establish a connection to the 'world' MySQL database on the local machine.
conn= mysql.connector.connect(host='localhost', user='root', password='', database='world')
```

```python
# Query the database for all columns from 'City' where 'CountryCode' is 'IND'.
pd.read_sql_query("SELECT * FROM City WHERE CountryCode LIKE 'IND'", conn)
```

```python
# Query the database for all columns from 'Country' where 'LifeExpectancy' exceeds 60.
pd.read_sql_query("SELECT * FROM Country WHERE LifeExpectancy > 60 " , conn)
```

```python
# Fetch all records from the 'Countrylanguage' table into a pandas DataFrame.
pd.read_sql_query("SELECT * FROM Countrylanguage" , conn)
```
