# Bivariate Analysis

```python
# import libraries
import pandas as pd
import seaborn as sns
```

```python
# laod dataset from seaborn
tips = sns.load_dataset('tips')
tips.head()
```

```python
# load titanic dataset from seaborn
titanic = sns.load_dataset('titanic')
titanic.head()
```

```python
# load flight dataset from seaborn
flight = sns.load_dataset('flights')
flight.head()
```

```python
# load iris dataset from seaborn
iris= sns.load_dataset('iris')
iris.head()
```

# Scatterplot (Numerical - Numerical)

```python
sns.scatterplot(data=tips, x='total_bill', y='tip', hue='sex', style= tips['smoker'], size=tips['size'])
```

# Barplot (Numerical - Categorical)

```python
sns.barplot(data= titanic, x='pclass', y='age', edgecolor='black', palette='Set2', hue=titanic['sex'])
```

# Boxplot (Numerical - Categorical)

```python
sns.boxplot(data=titanic, x='sex', y='age', hue='survived')
```

# Distplot (Numerical - Categorical)

```python
sns.distplot(titanic[titanic['survived'] == 0] ['age'], hist=False)
sns.distplot(titanic[titanic['survived'] == 1] ['age'], hist=False)
```

# Heatmap (Categorical- categorical)

```python
# Create a cross-tabulation showing the count of survivors for each passenger class
pd.crosstab(titanic['pclass'], titanic['survived'])
```

```python
sns.heatmap(pd.crosstab(titanic['pclass'], titanic['survived']))
```

```python
(titanic.groupby('pclass')['survived'].mean() *100).plot(kind='bar')
```

```python
titanic.head()
```

```python
titanic.groupby('embark_town')['survived'].mean() *100
```

# Clustermap (Categorical - Categorical)

```python
pd.crosstab(titanic['sibsp'], titanic['survived'])
```

```python
sns.clustermap(pd.crosstab(titanic['sibsp'], titanic['survived']))
```

# Pairplot

```python
sns.pairplot(titanic, hue='survived')
```

```python
sns.pairplot(iris, hue='species')
```

# Lineplot (Numerical - Numerical)

```python
flight.head()
```

```python
flight['year'] = flight['year'].astype(int)
flight.groupby('year').sum(numeric_only=True).reset_index()
flight.groupby('year').size().reset_index(name='count')


```

```python
# Sum passengers for each year
new =flight.groupby('year')['passengers'].sum().reset_index()

```

```python
new =sns.lineplot(new)
```

```python
# Create a pivot table showing passengers for each month (rows) and year (columns)
flight.pivot_table(values='passengers', index='month', columns='year')
```

```python
(sns.clustermap(flight.pivot_table(values='passengers', index='month', columns='year')))
```
