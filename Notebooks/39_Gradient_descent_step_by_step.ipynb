# `Gradient Descent`

```python
# import libraries
from sklearn.datasets import make_regression
import numpy as np
```

```python
# Make a small regression dataset with 4 samples and 1 feature
X, y = make_regression(n_samples=4, n_features=1, n_informative=1, n_targets=1, noise=80, random_state=13)
```

```python
# Plot the data points on a scatter plot
import matplotlib.pyplot as plt
plt.scatter(X, y)
```

```python
# import linear regression
from sklearn.linear_model import LinearRegression
```

```python
# Create a LinearRegression model and train it on X and y
reg = LinearRegression()
reg.fit(X, y)
```

```python
# Show the slope (coefficient) learned by the model
reg.coef_
```

```python
# Show the intercept (b value) learned by the model
reg.intercept_
```

```python
# Plot the data points and draw the regression line
plt.scatter(X, y)
plt.plot(X, reg.predict(X), color='red')
```

```python
# lets apply gradient descent assumint slope is constant m=78.35
# lets assume initial intercept b=0
y_pred = ((78.35* X) + 0).reshape(4)
```

```python
# Plot the data points and two regression lines:
# - Red line: model prediction (OLS)
# - Green line: predictions with intercept b = 0
# Add legend and show the plot
plt.scatter(X, y)
plt.plot(X, reg.predict(X), color='red', label='OLS Prediction')
plt.plot(X, y_pred, color='green', label='b =0')
plt.legend()
plt.show()
```

```python
# Calculate the gradient of the loss with respect to b (intercept) 
# for m = 78.35 and b = 0
m = 78.35
b = 0

loss_slope = -2 * np.sum(y - m * X.ravel() - b)
loss_slope
```

```python
# lets take the learning rate as 0.1
lr = 0.1

step_size = lr * loss_slope
step_size
```

```python
# calculate new intercept
b=b - step_size
b
```

```python
# Plot the data points and three lines:
# - Red: original OLS regression prediction
# - Green: prediction with updated b
# - Orange: prediction with b = 0
y_pred1 = ((78.35 * X) + b).reshape(4)

plt.scatter(X, y)
plt.plot(X, reg.predict(X), color='red', label='OLS Prediction')
plt.plot(X, y_pred1, color='green', label='b updated')
plt.plot(X, y_pred, color='orange', label='b =0')
plt.legend()
plt.show()
```

```python
# iteration 2
loss_slope = -2 * np.sum(y-m*X.ravel()-b)
loss_slope
```

```python
# step size
step_size = lr * loss_slope
step_size
```

```python
# Update the intercept b by subtracting the step size
b = b - step_size
b
```

```python
# Plot the data points and four lines:
# - Red: original OLS regression prediction
# - Green: prediction with updated b (2nd iteration)
# - Blue: prediction with updated b (1st iteration)
# - Orange: prediction with b = 0
y_pred2 = ((78.35* X) + b).reshape(4)
plt.scatter(X, y)
plt.plot(X, reg.predict(X), color='red', label='OLS Prediction')
plt.plot(X, y_pred2, color='green', label='b updated 2nd iter')
plt.plot(X, y_pred1, color='blue', label='b updated 1st iter')
plt.plot(X, y_pred, color='orange', label='b =0')
plt.legend()
plt.show()
```

```python
# iteration 3
loss_slope = -2 * np.sum(y-m*X.ravel()-b)
loss_slope
```

```python
# Calculate the step size for updating b using learning rate and gradient
step_size = lr * loss_slope
step_size
```

```python
# Update the intercept b by subtracting the step size
b= b - step_size
b
```

```python
# Plot the data points and four lines:
# - Red: original OLS regression prediction
# - Green: prediction with updated b (3rd iteration)
# - Blue: prediction with updated b (2nd iteration)
# - Orange: prediction with updated b (1st iteration)
y_pred = (m* X + b).reshape(4)
plt.scatter(X, y)
plt.plot(X, reg.predict(X), color='red', label='OLS Prediction')
plt.plot(X, y_pred, color='green', label='b updated 3rd iter')
plt.plot(X, y_pred2, color='blue', label='b updated 2nd iter')
plt.plot(X, y_pred1, color='orange', label='b updated 1st iter')
plt.legend()
plt.show()
```
