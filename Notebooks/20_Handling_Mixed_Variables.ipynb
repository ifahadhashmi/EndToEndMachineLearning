# Handling Mixed Variables

```python
# import libraries
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
# load dataset
df = pd.read_csv('../Dataset/titanic.csv')
df.head()
```

```python
# check unique values in 'number' column
df['number'].unique()
```

```python
# plot value counts of 'number' column
fig = df['number'].value_counts().plot.bar()
fig.set_title('passengers travelling with')
```

```python
# extract numerical part
df['number_numerical'] = pd.to_numeric(df['number'], errors = 'coerce', downcast='integer')
```

`coerce`: This is the most common choice for data cleaning. If it finds a non-numeric value, it will force it to become NaN (Not a Number). This allows the rest of your column to be converted to a numeric type, and you can handle the missing values later.

```python
# extract categorical part
df['number_categorical'] = np.where(df['number_numerical'].isnull(), df['number'], np.nan)
df.head()
```

ðŸ‘‰ If number_numerical is empty (NaN), then take the original value from number.

ðŸ‘‰ Otherwise, just leave it as NaN.

```python
# check unique values in 'Cabin' column
df['Cabin'].unique()
```

```python
# check unique values in 'Ticket' column
df['Ticket'].unique()
```

```python
df['cabin_num'] = df['Cabin'].str.extract('(\d+)') # capture numerical part
df['cabin_cat'] = df['Cabin'].str[0] # capture first letter
df.head()
```

```python
# plot value counts of 'cabin_cat' column
df['cabin_cat'].value_counts().plot(kind='bar')
```

```python
# extract the last bit of ticket as number
df['ticket_num'] = df['Ticket'].astype(str).apply(lambda s: s.split()[-1])
df['ticket_num'] = pd.to_numeric(df['ticket_num'], errors='coerce', downcast='integer')

# extract the first part of ticket as category
df['ticket_cat'] = df['Ticket'].astype(str).apply(lambda s: s.split()[0])
df['ticket_cat'] = np.where(df['ticket_cat'].str.isdigit(), np.nan, df['ticket_cat'])

df.head(15)
```
