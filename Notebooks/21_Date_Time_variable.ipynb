# working with dates

```python
# converting to datetime datatype
date['date'] = pd.to_datetime(date['date'])
```

```python
# get info about date dataset
date.info()
```

### Extract year

```python
date['date_year'] = date['date'].dt.year
date.sample(5)
```

### Extract month

```python
# extract month number from date
date['date_month_no'] = date['date'].dt.month
date.sample(5)
```

```python
# extract month name from date
date['date_month_name'] = date['date'].dt.month_name()
date.sample(5)
```

### Extract Days

```python
# extract day from date
date['date_day'] = date['date'].dt.day
date.sample(5)
```

```python
# day of week
date['date_dow'] = date['date'].dt.dayofweek
date.sample(5)
```

```python
# day of week name
date['date_dow_name'] = date['date'].dt.day_name()
date.drop(columns=['product_id', 'city_id', 'orders']).head()
```

```python
# is weekend?
date['date_is_weekend'] = np.where(date['date_dow_name'].isin(['Sunday', 'Saturday']), 1,0)
date.drop(columns=['product_id', 'city_id', 'orders']).head()
```

```python
# Extract the quarter from the 'date' column and create a new 'quarter' column
date['quarter'] = date['date'].dt.quarter

# Drop unnecessary columns: 'product_id', 'city_id', 'orders' and display the first few rows
date.drop(columns=['product_id', 'city_id', 'orders']).head()
```

### Extract semester

```python
# Create a 'semester' column: 1 for quarters 1 & 2, 2 for quarters 3 & 4
date['semester'] = np.where(date['quarter'].isin([1, 2]), 1, 2)

# Drop unnecessary columns and display the first few rows
date.drop(columns=['product_id', 'city_id', 'orders']).head()
```

### Extract time elapsed b/w dates

```python
# import library
import datetime

# get today's date
today=datetime.datetime.today()
today
```

```python
# calculate difference between today and date column
today -date['date']
```

```python
# calculate difference in days between today and date column
(today -date['date']).dt.days
```

```python
# Calculate the difference in years and multiply by 12, then add the difference in months
date['months_passed'] = (today.year - date['date'].dt.year) * 12 + (today.month - date['date'].dt.month)

# You can then see the result
date.head()
```
