# Univariate Analysis

```python
# import libraries
import pandas as pd
import seaborn as sns
```

```python
# load dataset from seaborn
df = sns.load_dataset('titanic')
df.head()
```

# Categorical Data

### Count Plot

```python
# Create a count plot showing the number of survivors (1) and non-survivors (0)
sns.countplot(df, x='survived')

# Display the exact count of each category in the 'survived' column
df['survived'].value_counts()
```

```python
# Define custom colors for each class (1st, 2nd, 3rd)
palette = ['red', 'green', 'blue']

# Create a count plot for passenger classes with custom colors and black borders
sns.countplot(df, x='pclass', palette=palette, edgecolor='black')
```

```python
# Define custom colors for male and female categories
palette = ['red', 'green']

# Create a count plot for gender distribution with custom colors and black borders
sns.countplot(df, x='sex', palette=palette, edgecolor='black')
```

```python
# Define custom colors for each embarkation port (C, Q, S)
palette = ['red', 'green', 'yellow']

# Create a count plot for embarkation points with custom colors and black borders
sns.countplot(df, x='embarked', palette=palette, edgecolor='black')
```

### Pie Chart

```python
# Plot a pie chart showing the percentage of passengers who survived vs. not survived
df['survived'].value_counts().plot.pie(autopct='%1.1f%%')
```

```python
# Plot a pie chart showing the percentage distribution of passengers across different classes (Pclass)
df['pclass'].value_counts().plot.pie(autopct='%1.1f%%')
```

```python
# autopct='%.2f' to show percentage on pie chart
df['sex'].value_counts().plot.pie(autopct='%1.1f%%') 
```

# Numerical Data

### Histogram

```python
import matplotlib.pyplot as plt
# Plot a histogram to visualize the age distribution of passengers with 30 bins
plt.hist(df['age'], bins=30)
```

### Distplot

```python
# Plot the distribution of 'age' with histogram and KDE (smooth curve)
sns.histplot(df['age'], kde=True)
```

### Boxplot

```python
# Create a boxplot to visualize the distribution and outliers in the 'age' column
sns.boxplot(df, x='age')
```

```python
# Find and display the minimum value in the 'age' column
df['age'].min()
```

```python
# Find and display the maximum value in the 'age' column
df['age'].max()
```

```python
# Calculate and display the mean (average) value of the 'age' column
df['age'].mean()
```

```python
# Calculate and display the skewness of the 'age' column 
# (shows how much the data is skewed  positive = right-skewed, negative = left-skewed)
df['age'].skew()
```

The age distribution is slightly right-skewed (skewness  0.39), indicating a few older passengers, but overall the ages are fairly symmetrically distributed.
